# Paroi_contact_VEF : diffusion_anisotrope

The goal of this notebook is to validate the implementation of the anisotropic diffusion used in the class Champ_front_contact_VEF. We compare 2 data sets, in 2D, with a uniform operator and an anisotropic one behaving like the first (diagonal matrix with equal values). The results are identical !

```python
from trustutils import run

run.introduction("Elie Saikali & Romain Le Tellier")
```

```python
run.reset()
run.initBuildDirectory()

cas = [ ("uniform" , "1 0.01") , ("aniso" , "4 0.01 0.0 0.0 0.01") ]

for c, val in cas:   
    run.addCaseFromTemplate("input.data", f"{c}", { "vals" : val} )
        
run.printCases()
run.runCases()
run.tablePerf()
```

# Convergence & time step evolution (in seconds)

```python
from trustutils import plot
  
a = plot.Graph(label_size=16, title_size=24, legend_size=16, nY=2)

a.addPlot(0,"Convergence")
for c, _ in cas: 
    a.addResidu(f"{run.BUILD_DIRECTORY}/{c}/input.dt_ev", label=f"{c}")
    a.scale(yscale="log")

a.addPlot(1,"dt_evol")
for c, _ in cas: 
    data = plot.loadText(f"{run.BUILD_DIRECTORY}/{c}/input.dt_ev")
    a.add(data[0], data[1], label=f"{c}")
    a.label("time","dt")
```

# Profiles

```python
   
a = plot.Graph(label_size=16, title_size=24, legend_size=16, nY=2)

a.addPlot(0,"T_moyen Solid 1")
for c, _ in cas: 
    data = plot.loadText(f"{run.BUILD_DIRECTORY}/{c}/input_T_SOLID1.son")
    a.add(data[0, 1:], data[1, 1:], label=f"{c}")
    a.label("time","T_moyen")

a.addPlot(1,"T_moyen Solid 2")
for c, _ in cas: 
    data = plot.loadText(f"{run.BUILD_DIRECTORY}/{c}/input_T_SOLID2.son")
    a.add(data[0, 1:], data[1, 1:], label=f"{c}")
    a.label("time","T_moyen")
```
